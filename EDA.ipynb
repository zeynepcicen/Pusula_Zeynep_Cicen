# Detect any anomalies or missing data

```python
 #TedaviSuresi ve UygulamaSuresi degiskenlerini sayıya çevirme
num_cols = ["TedaviSuresi", "UygulamaSuresi"]

for col in num_cols:
    df[col] = (
        df[col]
        .astype(str)                       # stringe çevir
        .str.extract(r'(\d+)')             # sadece sayıyı al
        .astype(float)                     # sayıya çevir
        .astype("Int64")                   # tam sayı yap (NaN destekler)
    )
```

```python
# Eksik verileri ısı haritası ile gösterelim

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='PuRd')
plt.title("Eksik Veri Kontrolü")
plt.show()
```

# To uncover patterns and relationships within the data

```python
df.head()
```

```python
# Pivot tablo: Bölüm x Yaş
import matplotlib.pyplot as plt
import seaborn as sns

pivot = df.pivot_table(index='Bolum', columns='Yas', aggfunc='size', fill_value=0)

plt.figure(figsize=(50,8))
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Yaş Kategorisine Göre Bölüm Başvuruları')
plt.ylabel('Bölüm')
plt.xlabel('Yaş Kategorisi')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(12,6))
sns.countplot(x='Bolum', hue='Cinsiyet', data=df, palette='autumn')
plt.xticks(rotation=45, ha='right')
plt.title('Cinsiyete Göre Başvurulan Bölüm')
plt.ylabel('Hasta Sayısı')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(data=df, x="Yas", hue="Cinsiyet", bins=20, kde=True, element="step")
plt.title("Yaşa Göre Cinsiyet Dağılımı")
plt.xlabel("Yaş")
plt.ylabel("Kişi Sayısı")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x="Yas", y="Cinsiyet", hue="Cinsiyet", dodge=True, alpha=0.7, size=3)
plt.title("Yaşa Göre Cinsiyet Dağılımı")
plt.xlabel("Yaş")
plt.ylabel("Cinsiyet")
plt.show()
```
