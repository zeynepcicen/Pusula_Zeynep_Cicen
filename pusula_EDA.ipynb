# Data visualization

```python
#Histograms
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Talent_Academy_Case_DT_2025.xlsx")

# 1. Yaş dağılım histogramı
plt.figure(figsize=(8,5))
sns.histplot(df['Yas'], bins=10, kde=True, color='skyblue')
plt.title("Hastaların Yaş Dağılımı")
plt.xlabel("Yaş")
plt.ylabel("Hasta Sayısı")
plt.show()

# 2. Cinsiyet dağılımı (bar plot)
plt.figure(figsize=(6,4))
sns.countplot(x='Cinsiyet', data=df, palette='pastel')
plt.title("Cinsiyet Dağılımı")
plt.xlabel("Cinsiyet")
plt.ylabel("Hasta Sayısı")
plt.show()

# 3. Tanılarla yaş dağılımı (boxplot)
plt.figure(figsize=(50,6))
sns.boxplot(x='Tanilar', y='Yas', data=df)
plt.xticks(rotation=90)  # tanı isimleri uzun ise döndür
plt.title("Tanılara Göre Yaş Dağılımı")
plt.xlabel("Tanılar")
plt.ylabel("Yaş")
plt.show()

# 4. Cinsiyet ve Tanılar ilişkisi (heatmap)
tanilar_cinsiyet = pd.crosstab(df['Tanilar'], df['Cinsiyet'])
plt.figure(figsize=(12,6))
sns.heatmap(tanilar_cinsiyet, annot=True, fmt='d', cmap='YlGnBu')
plt.title("Tanılar ve Cinsiyet İlişkisi")
plt.xlabel("Cinsiyet")
plt.ylabel("Tanılar")
plt.show()
```

```python
# Alerji boş değerlerini "Yok" olarak dolduralım

df['Alerji'] = df['Alerji'].fillna('Yok')

# Bar plot (Kan grubu ve Alerji dağılımı)
plt.figure(figsize=(10,6))
sns.countplot(x='KanGrubu', hue='Alerji', data=df, palette='Set2')
plt.title("Kan Grubu ve Alerji Dağılımı")
plt.xlabel("Kan Grubu")
plt.ylabel("Hasta Sayısı")
plt.legend(title='Alerji')
plt.show()

```

```python
import pandas as pd

# Excel dosyasını oku
data = pd.read_excel("Talent_Academy_Case_DT_2025.xlsx")

# --- KronikHastalik sütunu ---
kronik_split = data['KronikHastalik'].str.split(',', expand=True)
kronik_split.columns = [f'KronikHastalik_{i+1}' for i in range(kronik_split.shape[1])]
data = data.drop('KronikHastalik', axis=1)
data = pd.concat([data, kronik_split], axis=1)

# --- Bolum sütunu ---
bolum_split = data['Bolum'].str.split(',', expand=True)
bolum_split.columns = [f'Bolum_{i+1}' for i in range(bolum_split.shape[1])]
data = data.drop('Bolum', axis=1)
data = pd.concat([data, bolum_split], axis=1)

# --- Tedavi adi ---

tedavi_split = data['TedaviAdi'].str.split('\+', expand=True)
tedavi_split.columns = [f'TedaviAdi_{i+1}' for i in range(tedavi_split.shape[1])]
data = data.drop('TedaviAdi', axis=1)
data = pd.concat([data, tedavi_split], axis=1)

# --- Tanilar ---

tani_split = data['Tanilar'].str.split(',', expand=True)
tani_split.columns = [f'Tanilar_{i+1}' for i in range(tani_split.shape[1])]
data = data.drop('Tanilar', axis=1)
data = pd.concat([data, tani_split], axis=1)

# --- Alerji ---

alerji_split = data['Alerji'].str.split(',', expand=True)
alerji_split.columns = [f'Alerji_{i+1}' for i in range(alerji_split.shape[1])]
data = data.drop('Alerji', axis=1)
data = pd.concat([data, alerji_split], axis=1)



# Sonuç

data.head()
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np

# Excel dosyasını oku
df = pd.read_excel("Talent_Academy_Case_DT_2025.xlsx")

# 1. Eksik değerleri işleme
# Sayısal sütunlar için SimpleImputer (ortalama ile doldurma)
num_cols = ['Yas', 'UygulamaSuresi', 'TedaviSuresi']  # Süreleri sayısal yapacağız
df['UygulamaSuresi'] = df['UygulamaSuresi'].str.replace(' Dakika','').astype(float)
df['TedaviSuresi'] = df['TedaviSuresi'].str.replace(' Seans','').astype(float)

num_imputer = SimpleImputer(strategy='mean')
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Kategorik sütunlar için SimpleImputer (en sık görülen değer ile doldurma)
cat_cols = ['Cinsiyet', 'KanGrubu', 'Uyruk', 'Bolum']
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# 2. Kategorik değişkenleri kodlama
# OneHotEncoder kullanacağız
ohe_cols = [ 'Uyruk', 'Bolum']
df_encoded = pd.get_dummies(df, columns=ohe_cols)


# 3. Çoklu değerleri ayırma (KronikHastalik ve Tanilar)
# Örnek: ',' ile ayrılmış hastalıkları listeye çevirme
df_encoded['KronikHastalik'] = df['KronikHastalik'].fillna('').apply(lambda x: [i.strip() for i in x.split(',')] if x else [])
df_encoded['Tanilar'] = df['Tanilar'].fillna('').apply(lambda x: [i.strip() for i in x.split(',')] if x else [])

# Artık veri modellemeye hazır
df_encoded.head()
```
