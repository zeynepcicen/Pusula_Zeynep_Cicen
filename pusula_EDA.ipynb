# Detect any anomalies or missing data

```python
#Eksik verileri kontrol et
df.isnull().sum()
```

```python
# Eksik verileri ısı haritası ile gösterelim

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='PuRd')
plt.title("Eksik Veri Kontrolü")
plt.show()
```

# Data visualization

```python
#Histograms

import matplotlib.pyplot as plt

# 'Yas' değişkeninin histogramı
plt.figure(figsize=(8,5))
df['Yas'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Yaş Dağılımı')
plt.xlabel('Yaş')
plt.ylabel('Frekans')
plt.show()
```

```python
#Tedavi süresini sayısala çevirme

# '5 Seans' → 5 olarak dönüştür
df['TedaviSuresi_Sayisal'] = df['TedaviSuresi'].str.extract('(\d+)').astype(float)

#Yaş ve Tedavi Süresi arasındaki ilişki (scatter plot)

plt.figure(figsize=(8,5))
sns.scatterplot(x='Yas', y='TedaviSuresi_Sayisal', hue='Cinsiyet', data=df, palette='pastel')
plt.title('Yaş ve Tedavi Süresi İlişkisi')
plt.xlabel('Yaş')
plt.ylabel('Tedavi Süresi (Seans)')
plt.show()

#Hastalık ve tedavi süresi ilişkisi (boxplot)

plt.figure(figsize=(25,6))
sns.boxplot(x='KronikHastalik', y='TedaviSuresi_Sayisal', data=df)
plt.xticks(rotation=90)
plt.title('Hastalık ve Tedavi Süresi Dağılımı')
plt.xlabel('Hastalık')
plt.ylabel('Tedavi Süresi (Seans)')
plt.show()
```
